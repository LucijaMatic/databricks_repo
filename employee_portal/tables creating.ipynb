```python
%sql
CREATE OR REPLACE TABLE ekp_lm.employee_portal.document_metadata (
  document_id STRING PRIMARY KEY,
  document_name STRING NOT NULL,
  document_type STRING NOT NULL,
  source_weight FLOAT DEFAULT 1.0,
  file_format STRING,
  upload_date TIMESTAMP,
  uploaded_by STRING,
  file_size_bytes BIGINT,
  chunk_count INT,
  enabled BOOLEAN DEFAULT TRUE,
  source_url STRING,
  metadata_json STRING
) TBLPROPERTIES ("delta.feature.allowColumnDefaults" = "supported");
```

```python
%sql
CREATE OR REPLACE TABLE ekp_lm.employee_portal.document_chunks (
  chunk_id STRING PRIMARY KEY,
  document_id STRING NOT NULL,
  chunk_index INT,
  chunk_text STRING NOT NULL,
  chunk_embedding ARRAY<FLOAT>,
  token_count INT,
  FOREIGN KEY (document_id) REFERENCES ekp_lm.employee_portal.document_metadata(document_id)
) TBLPROPERTIES ("delta.feature.allowColumnDefaults" = "supported");


```

```python
%sql
ALTER TABLE ekp_lm.employee_portal.document_chunks SET TBLPROPERTIES (delta.enableChangeDataFeed = true);
```
